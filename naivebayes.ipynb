```python
import pandas as pd
train_df = pd.read_csv("data/poem_train.csv", encoding="utf-8")
test_df = pd.read_csv("data/poem_test.csv", encoding="utf-8")
```

```python
train_df
```

```python
test_df
```

```python
cs = train_df["作者"].astype("category").cat.categories
saved_map = {c:cs.get_loc(c) for c in cs}
saved_map
```

```python
train_df["作者"] = train_df["作者"].astype("category").cat.codes
train_df = train_df.drop(["詩名"], axis = 1)
train_df
```

```python
test_df["作者"] =test_df["作者"].replace(saved_map)
test_df = test_df.drop(["詩名"], axis = 1)
test_df
```

```python
import jieba 
def mytransform(data):
    return " ".join(jieba.cut(data))
train_df["內容"] = train_df["內容"].apply(mytransform)
train_df["內容"] = train_df["內容"].str.replace("\r", "")
train_df["內容"] = train_df["內容"].str.replace("\n", "")
train_df
```

```python
test_df["內容"] = test_df["內容"].apply(mytransform)
test_df["內容"] = test_df["內容"].str.replace("\r", "")
test_df["內容"] = test_df["內容"].str.replace("\n", "")
test_df
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vec = TfidfVectorizer()
bag = vec.fit_transform(train_df["內容"])
# for b in bag:
#     print(b)
```

```python
test_bag = vec.transform(test_df["內容"])
```

```python
print("特徵標籤:", vec.get_feature_names())
print("特徵維度:", len(vec.get_feature_names()))
```

```python
from sklearn.naive_bayes import MultinomialNB
clf = MultinomialNB(alpha = 0.001)
clf.fit(bag, train_df["作者"])
pre = clf.predict(test_bag)
print("預測結果:", list(pre))
print("正確結果:", list(test_df["作者"]))
```

```python
from sklearn.metrics import accuracy_score
print("正確率", accuracy_score(pre, test_df["作者"]))
```

```python
from sklearn.metrics import confusion_matrix
pd.DataFrame(confusion_matrix(test_df["作者"], pre))
```

```python
from sklearn.neighbors import KNeighborsClassifier
clf = KNeighborsClassifier(n_neighbors = 6)
clf.fit(bag, train_df["作者"])
pre = clf.predict(test_bag)
print("預測結果:", list(pre))
print("正確結果:", list(test_df["作者"]))
print("正確率", accuracy_score(pre, test_df["作者"]))
```
